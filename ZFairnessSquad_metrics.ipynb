```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from datasets import load_dataset
dataset = load_dataset("rajpurkar/squad_v2")

train_dataset = dataset['train']

nlp = spacy.load("en_core_web_sm")

# Function to check if the answer involves a named entity
def contains_named_entity(answer_text):
    doc = nlp(answer_text)
    return any(ent.label_ in ['PERSON', 'NORP', 'FAC', 'ORG', 'GPE', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW', 'LANGUAGE'] for ent in doc.ents)

questions = train_dataset['question']
answers = [ans['text'][0] if ans['text'] else 'No answer provided' for ans in train_dataset['answers']]
labels = [contains_named_entity(answer) for answer in answers]


vectorizer = CountVectorizer()
X = vectorizer.fit_transform(questions)


X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Training a logistic regression model
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy}")
```

```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from datasets import load_dataset
dataset = load_dataset("rajpurkar/squad_v2")

train_dataset = dataset['train']
# Load SpaCy for NER
nlp = spacy.load("en_core_web_sm")

def contains_named_entity(answer_text):
    doc = nlp(answer_text)
    return any(ent.label_ in ['PERSON', 'NORP', 'FAC', 'ORG', 'GPE', 'LOC', 'PRODUCT', 'EVENT', 'WORK_OF_ART', 'LAW', 'LANGUAGE'] for ent in doc.ents)


questions = train_dataset['question']
answers = [ans['text'][0] if ans['text'] else 'No answer provided' for ans in train_dataset['answers']]
labels = [contains_named_entity(answer) for answer in answers]


vectorizer = CountVectorizer()
X = vectorizer.fit_transform(questions)


X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)


model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy}")
```

```python
import numpy as np
arts_humanities_keywords = ['Beyoncé', 'Frédéric_Chopin', 'Sino-Tibetan_relations_during_the_Ming_dynasty', 'The_Legend_of_Zelda:Twilight_Princess', 'Spectre(2015_film)', 'To_Kill_a_Mockingbird', 'Kanye_West', 'Buddhism',]
science_technology_keywords = ['IPod', 'Genome', 'Internet', 'Comprehensive_school', 'Internet_service_provider', 'Computer', 'MP3', 'Computer_security']

def categorize_title(title):
    if any(keyword in title for keyword in arts_humanities_keywords):
        return 'Arts and Humanities'
    elif any(keyword in title for keyword in science_technology_keywords):
        return 'Science and Technology'
    else:
        return 'Other'
categories = [categorize_title(item['title']) for item in train_dataset]
_, _, _, cat_test = train_test_split(X, categories, test_size=0.2, random_state=42)

# Calculate SPD for 'Arts and Humanities' vs 'Science and Technology'
arts_indices = np.array(cat_test) == 'Arts and Humanities'
tech_indices = np.array(cat_test) == 'Science and Technology'

arts_rate = y_pred[arts_indices].mean()
tech_rate = y_pred[tech_indices].mean()

spd = tech_rate - arts_rate
print(f'Statistical Parity Difference (SPD): {spd}')
```

```python
import numpy as np

def calculate_theil_index(predictions, categories):

    arts_predictions = predictions[categories == 'Arts and Humanities']
    tech_predictions = predictions[categories == 'Science and Technology']

    combined = np.concatenate([arts_predictions, tech_predictions])

    total_mean = np.mean(combined)

    group_shares = [np.mean(group) / total_mean for group in [arts_predictions, tech_predictions]]

    proportions = [len(group) / len(combined) for group in [arts_predictions, tech_predictions]]

    # Calculating Theil index
    theil_index = np.sum([p * np.log(p / share) for p, share in zip(proportions, group_shares)])
    return theil_index

cat_test_array = np.array(cat_test)
theil_index = calculate_theil_index(y_pred, cat_test_array)
print(f'Theil Index: {theil_index}')
```
